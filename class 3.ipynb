# Linked list
in general, there are two varities of linked list, these are:
1. Singly-linked lists also knowm as "uni-directional" lists.
2. Doubly-linked lists also known as "bi-directional" lists.

unlike stacks and queues, linked lists allow for the addition or removal of nodes from anywhere in the collection. This means one can replace the 'head' of the list, insert anywhere in the middle of the list, as well as append to the list (which replaces the 'tail').

```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, datum):
            self.data = datum
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.tail:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        self.count += 1
        prev = None

        if not self.head:
            self.head = new_node
            self.tail = new_node
        
        elif index <= 0:
            new_node.next = self.head
            self.head = new_node

        current = self.head
        current_index = 0

        while current and current_index < index:
            prev = current
            current = current.next
            current_index += 1

        prev.next = new_node
        new_node.next = current

        if not current:
            self.tail = new_node

    def remove(self, value):
        current = self.head
        prev = None
        found = False
        while current and not found:
            if current.data == value:
                found = True
            else:
                prev = current
                current = current.next
        if found:
            if prev:
                prev.next = current.next
                if not prev.next:
                    self.tail = prev
            else:
                self.head = self.head.next
                if not self.head:
                    self.tail = None
        else:
            raise ValueError("Value not found")

    def index(self, value):
        current = self.head
        index = 0
        while current:
            if current.data == value:
                return index
            current = current.next
            index += 1
        raise ValueError(f"{value} is not in list")

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out

    def __len__(self):
        return self.count


```

```python
class DoublyLinkedList:
    class __Node:
        def __init__(self, datum):
            self.data = datum
            self.next = None
            self.prev = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def append(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.tail:  # Empty list
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        self.count += 1
        
        if not self.head:
            self.head = new_node
            self.tail = new_node
        
        elif index <= 0:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        
        current = self.head
        current_index = 0
        
        while current and current_index < index:
            current = current.next
            current_index += 1
        
        if current: 
            new_node.prev = current.prev
            new_node.next = current
            current.prev.next = new_node
            current.prev = new_node
        else:  # Inserting at end
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def remove(self, value):
        current = self.head
        found = False
        
        while current and not found:
            if current.data == value:
                found = True
            else:
                current = current.next
        
        if not found:
            raise ValueError("Value not found")
        
        self.count -= 1
        
        if current.prev:  
            current.prev.next = current.next
        else:  # Removing head node
            self.head = current.next
        
        if current.next: 
            current.next.prev = current.prev
        else:  # Removing tail node
            self.tail = current.prev

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out

    def __len__(self):
        return self.count
```

```python
dll = DoublyLinkedList()
dll.append(10)
dll.append(20)
dll.insert(1, 15)  
print(dll)         
dll.remove(15)
print(dll)        
print(len(dll))    
```
